```python
import json

import pandas as pd

from recommendations_manager import obtain_recommendations_randomanime
from main import anime_data
```

```python
codes = [int(entry.get("anime_id")) for entry in anime_data]
```

```python
algorithms = ["review_similarity", "synopsis_similarity", "character_similarity",
              "soft_clustering_new", "soft_clustering_popularity", "genre_similarity"]
```

```python
output = dict()
```

```python
for code in codes:
    recommendations = dict()
    for algo in algorithms:
        recommendation = obtain_recommendations_randomanime(code, algo, 1, 100)
        recommendation.update({"random_anime_code": list(map(lambda x: int(x), recommendation.get("random_anime_code", [])))})
        recommendations.update({algo:recommendation})
    output.update({code: recommendations})
```

```python
with open("freezed_recommendations.json", "w", encoding="utf-8") as f:    
    json.dump(output, f, indent=1)
    f.close()
```
